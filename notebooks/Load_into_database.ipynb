```python
import pandas as pd
import sqlite3
```

```python
patients = pd.DataFrame({
    'patient_id': [1,2,3],
    'first_name': ['Marie', 'Raiden', 'Aisha'],
    'last_name': ['Cole', 'Turner', 'Davis'],
    'birthdate': ['1985-04-12', '1990-09-22', '1978-01-15'],
    'gender': ['Female', 'Male', 'Female']
})

diagnoses = pd.DataFrame({
    'diagnosis_code': ['D02', 'M01', 'V05'],
    'diagnosis_description': ['Flu', 'Hypertension', 'Diabetes']
})

visits = pd.DataFrame({
    'visit_id': [101, 102, 103, 104, 105],
    'patient_id': [1, 2, 2, 3, 3],
    'visit_date': ['2026-01-10', '2026-02-14', '2026-01-18', '2026-03-20', '2026-04-01'],
    'diagnosis_code': ['D02', 'D02', 'M01', 'V05', 'V05'],
    'visit_duration_minutes': [30, 25, 40, 35, 50]
})
```

```python
conn = sqlite3.connect('Health_Project.db')
```

```python
patients.to_sql('patients', conn, if_exists='replace', index=False)
diagnoses.to_sql('diagnoses', conn, if_exists='replace', index=False)
visits.to_sql('visits', conn, if_exists='replace', index=False);
```

```python
pd.read_sql('SELECT * FROM patients', conn)
```

```python
conn.close()
```
